```python
import pandas as pd # Libreria Pandas
import sqlalchemy as db 
```

```python
# Crear una conexion 
# Cadena: usuario, clave, host o servidor, puerto, nombre de la bd
engine = db.create_engine("mysql+mysqlconnector://root:root@127.0.0.1:3310/db_movies_netflix_transact")

#Establecemos op iniciamos la conexion
conn = engine.connect()
```

# Cargando Datos a Dimension Moview

```python
query = """
    SELECT 
        movie.movieID as movieID, movie.movieTitle as title, movie.releaseDate as releaseDate,
        gender.name as gender , person.name as participantName, participant.participantRole as roleparticipant
    FROM movie
    INNER JOIN participant
    ON movie.movieID=participant.movieID
    INNER JOIN person
    ON person.personID = participant.personID
    INNER JOIN movie_gender
    ON movie.movieID = movie_gender.movieID
    INNER JOIN gender
    ON movie_gender.genderID = gender.genderID
    """
```

```python
#Cargar los resultados de la consulta en DF
movies_data = pd.read_sql(query, con=conn)

#Transfromar a entero
movies_data["movieID"]=movies_data["movieID"].astype('int')

#Mostrar elementos
movies_data
```

```python
# Cargar un CSV
movies_award = pd.read_csv("./data/Awards_movie.csv")

# Transformar el id a entero
movies_award["movieID"] = movies_award["movieID"].astype('int')

# Renombrar la columna de premio
movies_award.rename(columns={"Aware": "Award"}, inplace=True)

movies_award
```

```python
# Cruzar (join) 
movies_data_final = pd.merge(movies_data,movies_award,
                                left_on="movieID", right_on="movieID"
)

movies_data_final
```

```python
# Crear una conexion a la BD WareHouse
enginedw = db.create_engine("mysql+mysqlconnector://root:root@127.0.0.1:3310/dw_netflix")

conn_dw = enginedw.connect()
```

```python
# Renombrar dos columnas
movies_data_final = movies_data_final.rename(columns={"releaseDate":"releaseMovie","Award":"awardMovie"})

# Eliminar un campo
movies_data_final = movies_data_final.drop(columns=['IdAward'])
```

```python
movies_data_final
```

```python
movies_data_final.to_sql('dimMovie', conn_dw, if_exists='append', index=False)
```

```python
whatchs_data = pd.merge(users_id,movies_id,how='cross')

whatchs_data
```

```python
import random
from datetime import datetime,timedelta

def gen_rating():
    numero_aleatorio = round(random.uniform(0,5),1)

    return numero_aleatorio

def gen_timestamp():
    start_date = datetime(2024, 1 , 15)
    end_date = datetime(2024, 4, 6)

    random_date = start_date + timedelta(seconds=random.randint(0,int((end_date - start_date).total_seconds())))

    return random_date

```

```python
whatchs_data['rating'] = whatchs_data['movieID'].apply(lambda x: gen_rating())

whatchs_data['timestamp'] = whatchs_data['userID'].apply(lambda x: gen_timestamp())


```
